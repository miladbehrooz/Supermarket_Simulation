### Load Libraries

```python
import os
import random
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pygraphviz as pgv
from datetime import datetime
from datetime import timedelta
from faker import Faker


sns.set()
plt.rcParams['figure.figsize'] = (12, 6)
```

### Data Preprocessing

#### Load data and concat dataframes

```python
files = os.listdir('../data/')
df_list = []
for file in files:
    print(file)
    df = pd.read_csv(f'../data/{file}',sep=';',parse_dates=True,index_col=0)
    df_list.append(df)
df_all = pd.concat(df_list,sort=True)
```

```python
df_all.head()
```

```python
df_all.info()
```

#### Set distinct ID for each customers

```python
df_all.sort_index(inplace=True)
df_all['hour'] = df_all.index.hour
df_all['day'] = df_all.index.day_name()
df_all['customer_id'] = df_all['day'].apply(lambda x: x[0:2]) + df_all['customer_no'].astype(str)
```

```python
df_all.tail(3)
```

#### Resample the timestamp in minutes

```python
def resample_timestamp(customer):
    """
    return the resampled dataframe in minute for the given customer_id
    """
    df_partial = df_all[df_all['customer_id'] == customer]
    df_partial_resampled = df_partial.resample('1min').min()
    
    df_partial_resampled .sort_index(inplace=True)
    df_partial_resampled ['time'] = df_partial_resampled .index.time
    df_partial_resampled ['hour'] = df_partial_resampled .index.hour
    df_partial_resampled ['day'] = df_partial_resampled .index.day_name()
    df_partial_resampled ['customer_no'].fillna(method="ffill", inplace=True)
    df_partial_resampled ['customer_no'].astype(int)
    df_partial_resampled ['customer_id'] = df_partial_resampled ['day'].apply(lambda x: x[0:2]) + df_partial_resampled ['customer_no'].astype(str)
    df_partial_resampled ['location'].fillna(method="ffill", inplace=True)
    
    return df_partial_resampled
    
```

```python
df_list =[]
for i,customer in enumerate(df_all['customer_id'].unique()):
    df_partial_resample =resample_timestamp(customer)
    df_list.append(df_partial_resample)
df_all_resampled = pd.concat(df_list,axis=0)
```

### Transition Propabilities

#### compute Transition probabilities

```python
df_all_resampled = df_all_resampled.sort_values(['timestamp','customer_no'])
df_all_resampled
```

```python
pre_crosstab = df_all_resampled.reset_index().sort_values(['customer_no', 'timestamp'])
```

```python
pre_crosstab['previous_location'] = pre_crosstab.groupby('customer_id')['location'].apply(lambda x: x.shift(1))
```

```python
pre_crosstab.fillna('entrance',inplace=True)
```

```python
pre_crosstab
```

```python
trans_prob = pd.crosstab(pre_crosstab['previous_location'], pre_crosstab['location'], normalize=0)
```

```python
trans_prob['entrance'] = [0,0,0,0,0] # add enterance columns to the transition probability matrix
trans_prob.loc['checkout'] = [1,0,0,0,0,0] # add checkout row to the transition probability matrix
```

```python
# sort the columns and rows of transiotin matrix
trans_prob_sorted = np.zeros((6,6))
for i,col in enumerate(trans_prob.columns):
    for j,row in enumerate(trans_prob.columns):
        trans_prob_sorted[i,j] = trans_prob.loc[col,row]

trans_prob_sorted = pd.DataFrame(trans_prob_sorted, columns = trans_prob.columns,index = trans_prob.columns).reset_index().rename(columns={'location':'previous_location'}).set_index('previous_location')
```

```python
trans_prob_sorted
```

```python
sns.heatmap(data=trans_prob_sorted,square=True, annot=True, cmap='Blues');
```

#### Draw state diagram

```python
locations = trans_prob_sorted.columns
G = pgv.AGraph(directed=True)
for start_location in locations[:-1]:
    for end_location in locations:
        G.add_edge(
            start_location,
            end_location,
            label=np.round(trans_prob.loc[start_location, end_location], 2))
G.draw("../figures/transition.png", prog="dot")
```

### MC Simulation

### MCMC Simulation

#### MCMC simulation for single customer

```python
class Customer():
    """
    a single customer that moves through the supermarket
    in a MCMC simulation
    """
    def __init__(self,name,state,trans_prob,customer_id=0):
        self.name = name
        self.state = state
        self.trans_prob = trans_prob
        self.id = customer_id
        
    def __repr__(self):
        return f'Customer {self.name} is in section {self.state}'
    
    def next_state(self):
        """
        Propagates the customer to the next state.
        Returns nothing.
        """
        self.state = np.random.choice(['checkout','dairy','drinks','fruit','spices','entrance'],
                                    p=list(trans_prob_sorted.loc[self.state]))
    def is_active(self):
        """Returns True if the customer has not reached the checkout yet.
        """
        if self.state != 'checkout':
            return True
        else:
            return False
```

```python
cust1 = Customer("Jake", "drinks",trans_prob_sorted)
```

```python
while True:
    if cust1.is_active() == False:
        print(f'Customer {cust1.name} left the supermarket.' )
        break
    cust1.next_state()
    print(cust1)
```

#### MCMC simulation for multiple customers

```python
class Supermarket:
    """manages multiple Customer instances that are currently in the market.
    """

    def __init__(self):        
        # a list of Customer objects
        self.customers = []
        self.data = pd.DataFrame(columns=['timestamp','customer_name','customer_no', 'location'])
        self.minutes = 0
        self.last_id = 0
        self.time = datetime.strptime('2019-09-06 07:00:00', "%Y-%m-%d %H:%M:%S")

    def __repr__(self):
        return f'The are {len(self.customers)} in the supermarket.'

    def get_time(self):
        """current time in HH:MM format,
        """
        current_time = datetime.now()
        return datetime.strptime(current_time,"%m-%d-%Y %H:%M:%S")


    def print_customers(self):
        """print all customers with the current time and id in CSV format.
        """
        for i in range(len(self.customers)):
            print(f'{self.time.strftime("%Y-%m-%d %H:%M:%S")}, customer {self.customers[i].name} with customer_no = {self.customers[i].id} is in {self.customers[i].state} section')
           
            self.data = self.data.append({'timestamp': self.time.strftime("%m-%d-%Y %H:%M"),
                        'customer_name': self.customers[i].name, 'customer_no': self.customers[i].id,
                        'location':self.customers[i].state},ignore_index=True)
        self.data.to_csv('../data/simulation.csv')
        return None

    def next_minute(self):
        """propagates all customers to the next state.
        """
        for i in range(len(self.customers)):
            self.customers[i].next_state()
            
        self.time += timedelta(minutes=1)
        return None
    
    def add_new_customers(self):
        """randomly creates new customers.
        """
        section = np.random.choice(['checkout','dairy','drinks','fruit','spices'])
        cust = Customer(Faker().first_name(), section ,trans_prob_sorted,customer_id=self.last_id)
        self.customers.append(cust)
        self.last_id += 1
        
        return None

    def remove_exitsting_customers(self):
        """removes every customer that is not active any more.
        """
        active = [x.is_active() for x in self.customers] 
        self.customers = [self.customers[i] for i in range(len(self.customers)) if active[i]]

        return None
```

```python
N = 5  # maximum number of customers who can enter to the supermarket per minute 
time = 14 * 60  # simulation time in minute
aldi = Supermarket() 
for t in range(time):
    for i in range(random.randint(0, N)):
        aldi.add_new_customers()
        
    aldi.print_customers()   
    aldi.next_minute()
    aldi.remove_exitsting_customers()
```
